```python
# Rango intercuartílico
(
    df
    .quantile(q=[0.75, 0.50, 0.25], numeric_only=True)
    .T
    # Cambianos de nombre al axis
    .rename_axis('Metric')
    .reset_index()
    #Creamos una nueva columna y restamos los cuartiles 75 y 25
    .assign(iqr=lambda df:df[0.75] - df[0.25])
)
```

```python
# Graficamos los caurtiles 25 y 75, ademas de las media y mediana

#Grafico de berras de la longitud del pico
hist = sns.histplot(data=df,
                    x='flipper_length_mm',
                    binwidth=1,
                    hue='species',
                    alpha=.5)
hist.set_title('Penguins Flipper Length Distribution', fontsize=15, y=1.03)

# Graficamos lineas vericales

#Media
plt.axvline(x=df.flipper_length_mm.mean(),
            linestyle='dashed',
            color='r',
            linewidth=2)

#Mediana
plt.axvline(x=df.flipper_length_mm.median(),
            linestyle='dashed',
            color='g',
            linewidth=2)

#Quartil 75
plt.axvline(x=df.flipper_length_mm.quantile(0.75),
            linestyle='dashed',
            color='k',
            linewidth=2)

#Quartil 25
plt.axvline(x=df.flipper_length_mm.quantile(0.25),
            linestyle='dashed',
            color='k',
            linewidth=2)


# Agregamos texto a las lineas verticales
plt.text(184.7, 17.5, 'Q(25%)', fontsize=12)
plt.text(192, 17.5, 'Median', fontsize=12, color='g')
plt.text(201.5, 17.5, 'Mean', fontsize=12, color='r')
plt.text(213.5, 17.5, 'Q(75%)', fontsize=12)

plt.show()
```

```python
#Filtramos con query las especies
adelie_df = df.query("species== 'Adelie'")
gento_df = df.query("species== 'Gentoo'")
chinstrap_df = df.query("species== 'Chinstrap'")

list_categories = [adelie_df, gento_df, chinstrap_df]
```

```python
# Otro metodo mas escalable
species_list = ['Adelie', 'Gentoo', 'Chinstrap']
list_categories = [df[df['species'] == specie] for specie in species_list]
```

```python
fig, ax = plt.subplots(1, len(list_categories), sharey=True)

for i, categories in enumerate(list_categories):
    sns.histplot(ax = ax[i],
                 data=categories,
                 x='flipper_length_mm',
                 multiple='layer',
                 hue='species',
                 bins=15,
                 kde=True,
                 palette=penguin_color)
    
plt.show()
```
