```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('./base.xlsx')
df1.head()
```

```python
df1 = df1[['SUMINISTRO','MARCA_RET', 'MODELO', 'SERIE_RET', 'SISTEMA']]
# df1['SUMINISTRO'] = df1['SUMINISTRO'].astype(str).str.strip()
# df1['SERIE_RET'] = df1['SERIE_RET'].astype(str).str.strip()
df1 = df1.astype(str).apply(lambda col: col.str.replace(' ', '', regex=False))
df1.dtypes
```
