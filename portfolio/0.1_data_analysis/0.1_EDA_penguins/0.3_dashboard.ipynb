```python
# Descargamos los datos
df = pd.read_csv('../../../data/processed/datos_procesados_penguins.csv')
```

```python
# Ajusta el tamaño de la figura para el dashboard
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
axes = axes.flatten()

# Gráfico 1: Conteo por especies
df['species'].value_counts().plot(kind="bar",
                                  color=penguin_color.values(),
                                  ax=axes[0])
axes[0].set_title("Conteo por especies")

# Grafico 2: Longitud del pico
sns.histplot(data = df,
             x='flipper_length_mm',
             binwidth=1,
             hue='species',
             alpha=.5, # Transparencia
             ax=axes[1])

# Gráfico 3: Relación entre longitud y profundidad del pico
sns.scatterplot(data=df, 
                x='bill_length_mm', 
                y='bill_depth_mm', 
                hue='species', 
                alpha=0.5, 
                s=100, 
                ax=axes[2])
axes[2].set_title("Relación entre longitud del pico y profundidad por especie")

#Gráfico 4: KDE por especies (longitud de ala vs peso)
sns.kdeplot(data=df, 
            x='body_mass_g', 
            y='flipper_length_mm', 
            hue='species', 
            palette=penguin_color, 
            ax=axes[3])
axes[3].set_title("Relación entre longitud del ala y peso corporal por especies")

# Gráfico 5: KDE por islas
sns.kdeplot(data=df, 
            x='body_mass_g', 
            y='flipper_length_mm', 
            hue='island', 
            palette=penguin_color, 
            ax=axes[4])
axes[4].set_title("Relación entre longitud del ala y peso corporal por islas")

# Gráfico 6: Stripplot por especies
sns.stripplot(data=df, 
              x='species', 
              y='flipper_length_mm', 
              hue='species', 
              palette=penguin_color, 
              dodge=True, 
              ax=axes[5])
axes[5].set_title("Especies por longitud del ala")

fig.suptitle("Análisis Exploratorio de Datos de Pingüinos", fontsize=20, fontweight='bold', y=1.02)



conclusion_text = """
CONCLUSIONES 

1. La especie Adelie es la más representada en el conjunto de datos.
2. Existen diferencias claras en la proporción de especies según las variables anatómicas.
3. El análisis de correlaciones muestra relaciones interesantes entre longitud de aletas y peso corporal.
4. Las islas influyen significativamente en la distribución de las especies.
"""

fig.text(0.5, -0.02, conclusion_text, fontsize=12, ha='center', va='top', wrap=True)



plt.tight_layout()
plt.show()
```
